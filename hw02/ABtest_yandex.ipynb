### A/B эксперимент. 

Необходимо определить минимальный размер экспериментальной выборки, чтобы  разница долей кликнутых 
поисковых выдач в 1% была бы статистически значима с p-value = 0.05. 

Исходные данные: 
   - размеры выборок(экспериментальной и контрольной) одинаковые и равны:
   n1 = n2 = n = 5% * 200млн = 10млн
   - минимальныйй размер эффекта $\delta$ = 0.01
   - p-value = 0.05
   - оцкнка конверсии в контрольной группе равна $ p_{1}^{\hat{}} = 0.1 $

 Чтобы ответить на вопрос сколько же нам нужно проводить по времени эксперимент, нужно понять как
мы будем оценивать результаты. Для этогo необходимо выбрать подходящий статистический тест. В 
данном примере мы сравниваем равенство долей в двух выборках, при этом можно считать, что выборки несвязные - разным пользователя показываются разные варианты поисковой выдачи.

В данном случае будем использовать t-test стьюдента для 2 несвязных выборок, так как реальную
дисперсию мы не знаем ( знаем только оценку на контрольной группе), а средние значения конверсий
в двух группах можно считать нормально распределенными по центральной предельной теореме. ( настоящее
распределение биномиальное). 

Теперь, чтобы оценить срок A/B теста нам нужно посчитать значение статистики на наших данных 
и сравнить с критическим значением распределения стьюдента - в данном случае это t-квантиль 
уровня $ \alpha = 0.95 $. Проблема в том, что нам ничего пока не известно про   $ p_{2}^{\hat{}} $ - среднее значение конверсии в экспериментальной группе. 

Будем пользоваться предположением, что выборочные дисперсии в обоих группах совпадают. Тогда подставим данные ( вместо разница средних - $\delta$ = 0.01 , а вместо выборочной дисперсии 0.1*0.9 = 0.09 ) и получим значение стаситики t: 

```python
import scipy.stats
import numpy as np
```

```python
def t_stat( delta, p, n):
    df = 2*n - 2 # degrees of fredom
    S = np.sqrt(p*(1-p)*(n-1)/n)
    return delta / (S*(2/np.sqrt(n))), df
```
