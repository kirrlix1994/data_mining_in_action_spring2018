Необходимо определить минимальный размер экспериментальной выборки, чтобы  разница долей кликнутых 
поисковых выдач в 1% была бы статистически значима с p-value = 0.05. 

Исходные данные: 
   - размеры выборок(экспериментальной и контрольной) одинаковые и равны:
   n1 = n2 = n = 5% * 200млн = 10млн
   - минимальныйй размер эффекта $\delta$ = 0.01
   - p-value = 0.05

 Чтобы ответить на вопрос сколько же нам нужно проводить по времени эксперимент, нужно понять как
мы будем оценивать результаты. Для этогo необходимо выбрать подходящий статистический тест. В 
данном примере мы сравниваем равенство долей в двух выборках, при этом можно считать, что выборки несвязные - разным пользователя показываются разные варианты поисковой выдачи.

Нам подоходит Z-критерий равеннства долей. Критерий предполагает, чтьо выборки имеют параметрическое 
распредление бернули и независимые.

Воспольземся центральной предельной теоремой для величины случайной величины 
$ p_{1}^{\hat{}} - p_{2}^{\hat{}}$.  Дейстчвительно, каждое из слагаемых - среднее значение суммы
независимых одинаково распределенных величин, поэтому слагаемые имеют нормальное распределение,
а сумма двух нормальных расперделений также имеет нормальное распределение.

Чтобы получить статистику Z ~ N(0,1)  имеющую нормальное распределение со средним 0 и дисперсией 1,
вычтем из нашей случайной величины $ p_{1}^{\hat{}} - p_{2}^{\hat{}} $  cреднее и поделим на 
стандартное отклонение.

  В данном случае наща гипотезу $ H0 : p_{1} - p_{2} = 0 $.  При проверке гипотез необходимо получить распределение статистики при условии, что гипотеза H0 верна. А тогда реальное среднее - 0. Дисперсия будет равна сумме димперсий(выборки независимые), и поэтому получаем статистику:

$ Z = p_{1}^{\hat{}} - p_{2}^{\hat{}} $

$$ Z = \dfrac{p_{1}^{\hat{}} - p_{2}^{\hat{}}} { \sqrt{ \dfrac{p_{1}^{\hat{}}(1 -p_{1}^{\hat{}})}{n} +   \dfrac{p_{2}^{\hat{}}(1 -p_{2}^{\hat{}})}{n}}} $$

$ Z = \dfrac{p_{1}^{\hat{}} - p_{2}^{\hat{}}}{ \dfrac{p_{1}^{\hat{}}(1 -p_{1}^{\hat{}})}{n}}$
